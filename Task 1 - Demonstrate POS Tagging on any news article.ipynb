```python
data_sen=dataset['Text'] # retrive Text column from dataset 
print(len(data_sen)) # length is 735
data_sen[0] #show the paragraph which is located index 0
```

# 2. Preprocessing on text

```python
from nltk.tokenize import word_tokenize, sent_tokenize  # import for word tokenizer and sentence tokenizer

import nltk # import nltk liberary
from nltk.corpus import stopwords # this is for remove stopwords from sentences


for_token_ = sent_tokenize(data_sen[0]) # for further POS operations we are going to use only index 0 paragraph and perform tokenization on sentences 
for_token_
```

```python
to=[] #declare list
for i in for_token_: 
        word_sen = nltk.word_tokenize(i) # here perform tokenization on words
        
        word_sen = [new_word for new_word in word_sen if not new_word in  stopwords.words('english')] # here we are checking stopwords if word  is stopword reject else assign into word_sen variable
        to.append(nltk.pos_tag(word_sen)) # if word is not stop word append into list
        
        

    
```

# Step 3 -display the POS tags with the Words.

```python
# here we are printing words and Pos realted to words
for i in range(0,len(to)):
    print("---------------------")

    print("sentence no {} Pos tags \n ".format(i+1))

    print ("{:<15} {}".format('Words', 'POS Tag')) 
    print("\n")
    for k, l in to[i]: 
         print ("{:<15} {}".format(k,l)) 
```
